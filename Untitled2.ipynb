```python
n = int(input("Enter number of elements: "))
a = []
print("Enter", n, "integers:")
for _ in range(n):
    x = int(input())
    a.append(x)
current_max = a[0]
current_min = a[0]
for i in range(1, n):
    if a[i] > current_max:
        current_max = a[i]
    if a[i] < current_min:
        current_min = a[i]

print("Maximum =", current_max)
print("Minimum =", current_min)
```

```python
def find_pair_with_diff(arr, x):
    arr.sort()
    n = len(arr)
    i, j = 0, 1
    while i < n and j < n:
        if i == j:
            j += 1
        else:
            diff = arr[j] - arr[i]

            if diff == x:
                return arr[i], arr[j]
            elif diff < x:
                j += 1
            else:
                i += 1
    return None
n = int(input("Enter number of elements: "))
nums = []
print("Enter", n, "integers:")
for _ in range(n):
    nums.append(int(input()))
x = int(input("Enter x (required absolute difference): "))
pair = find_pair_with_diff(nums, x)
if pair is not None:
    a, b = pair
    print("Pair with difference", x, "is:", a, b)
```
