```python
n = int(input("Enter number of elements: "))
arr = list(map(int, input("Enter the numbers (separated by spaces): ").split()))

minimum = arr[0]
maximum = arr[0]

for i in range(1, n):
    if arr[i] < minimum:
        minimum = arr[i]
    if arr[i] > maximum:
        maximum = arr[i]

print("Maximum =", maximum)
print("Minimum =", minimum)
```

```python
s=input("Enter String: ")
left, right = 0, len(s) - 1
while left < right:
    if s[left] != s[right]:
        print("Not palindrome")
        break
    left += 1
    right -= 1
else:
    print("Palindrome")
```

```python
n = int(input("Enter number of elements: "))
arr = list(map(int, input("Enter the elements: ").split()))
x = int(input("Enter x: "))

arr.sort()

left = 0
right = 1

while right < n:
    diff = arr[right] - arr[left]

    if diff == x:
        print(arr[left], arr[right])
        break
    elif diff < x:
        right += 1
    else:
        left += 1
        if left == right:
            right += 1
```

```python
n = int(input("Enter number of elements: "))
arr = list(map(int, input("Enter the array elements: ").split()))
a, b, c = map(int, input("Enter values of a, b, c: ").split())
low = 0
mid = 0
high = len(arr) - 1
while mid <= high:
    if arr[mid] == a:
        arr[low], arr[mid] = arr[mid], arr[low]
        low += 1
        mid += 1
    elif arr[mid] == b:
        mid += 1
    else:
        arr[mid], arr[high] = arr[high], arr[mid]
        high -= 1
print("Sorted array:", arr)
```
