```python
#Q1
n = int(input("Enter number of elements: "))
arr = []
for i in range(n):
    arr.append(int(input()))
min = max = arr[0]
for i in range(1, n):
    if arr[i] < min:
        min = arr[i]
    elif arr[i] > max:
        max = arr[i]
print("Minimum:", min)
print("Maximum:", max)
```

```python
#Q2
s = input("Enter string: ")
left, right = 0, len(s) - 1
while left < right:
    if s[left] != s[right]:
        print("Not a palindrome")
        break
    left += 1
    right -= 1
else:
    print("Palindrome")
```

```python
#Q3
n = int(input("Enter number of elements: "))
arr = []
for i in range(n):
    arr.append(int(input()))
x = int(input("Enter value of x: "))
arr.sort()
i = 0
j = 1
while j < n:
    diff = arr[j] - arr[i]
    if diff == x:
        print(arr[i], arr[j])
        break
    elif diff < x:
        j += 1
    else:
        i += 1
        if i == j:
            j += 1
```

```python
#Q4
n = int(input("Enter number of elements: "))
arr = []
for i in range(n):
    arr.append(int(input()))
a = int(input("Enter value of a: "))
b = int(input("Enter value of b: "))
c = int(input("Enter value of c: "))
if a > b:
    a, b = b, a
if b > c:
    b, c = c, b
if a > b:
    a, b = b, a
i = 0
mid = 0
j = n - 1
while mid <= j:
    if arr[mid] == a:
        arr[i], arr[mid] = arr[mid], arr[i]
        i += 1
        mid += 1
    elif arr[mid] == b:
        mid += 1
    else:
        arr[mid], arr[j] = arr[j], arr[mid]
        j -= 1
print("Sorted array:", arr)
```
