```python
arr = list(map(int, input().split())) # to input an array
max = arr[0]
min = arr[0]
for x in arr:
  if x>max: max = x
for x in arr:
  if x<min: min = x

print("\n")
print(arr)
print("Min:", min, " Max:", max)
```

QUESTION 3

```python
lst = list(map(int, input().split())) # to input an array
x = int(input("Enter the desired difference:"))

n = len(lst)
found = 0
lst.sort()

i, j = 0, 1

while i<n and j<n:
  diff = lst[j]-lst[i]
  if diff==x and i!=j:
    found = 1
    print("pair:", lst[i], ",", lst[j])
    break

  elif diff>x: i+=1
  elif diff<x: j+=1

if found==0: print("No such pair found")
```
